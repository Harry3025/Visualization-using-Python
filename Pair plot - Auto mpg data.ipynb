```python
# Import Neccesary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the given dataset

data = pd.read_csv('C:/Users/vicky/Downloads/auto_data.csv')
data.head()
```

```python
# Checking for unique categories in origin columns

data['origin'].unique()
```

```python
# 1 = America, 2= Europe, 3 = Asia

def origin(num):
    if num  == 1:
        return 'America'
    elif num == 2:
        return 'Europe'
    else:
        return 'Asia'
    
data['origin'] = data['origin'].apply(origin)
```

```python
# Pair plot

sns.pairplot(data[['mpg','weight','origin']],hue = 'origin', size = 4)
```
