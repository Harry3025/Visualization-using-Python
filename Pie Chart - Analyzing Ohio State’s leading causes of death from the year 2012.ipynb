```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# cause of deaths
cause = ['Chronic Disease', 'Unintetional Injuries', 'Alzheimers', 'Infuenza and Pneumonia', 'Sepsis', 'Others']

# Percentile (in order with cause of deaths as mentioned above)
percentile = [62,5,4,2,1,26]
```

```python
plt.figure(figsize = (8,8))
myexplode = [0.05,0,0,0,0,0]
plt.pie(percentile,labels = cause, autopct = '%1.1f%%', startangle = 90, explode = myexplode)
plt.title("Ohio State leading causes of death- Year 2012")
plt.show()
```
